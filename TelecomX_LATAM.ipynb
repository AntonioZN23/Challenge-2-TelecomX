```python
#Normalizar datos y actualizar DF
lista = df.columns[2:]
for j in lista:
  df_variable = pd.json_normalize(df[j])
  df = df.join(df_variable)
  df = df.drop(j,axis='columns')

df.head()
```

```python
df.info()
```

#ðŸ”§ TransformaciÃ³n

```python
df['Charges.Total'] = df[['Charges.Total']].map(lambda x: x.replace('$','').replace(',','').strip())
df['Charges.Total'] = pd.to_numeric(df['Charges.Total']).astype(np.float64)
```

```python
df.info()
```

```python
valores = list(df['Charges.Monthly']/30)
```

```python
df = df.join(pd.DataFrame(data=valores,columns=['Cuentas_Diarias'],index=df.index,dtype=(np.float64)))
df.head()
```

```python
#Se verifica la cantidad de espacios vacÃ­os y si es menor al 5% se elimina
columnas_si_a_numero = ['Churn','Partner','Dependents','PhoneService',]

for j in columnas_si_a_numero:
  df[j] = df[j].str.lower()
  df[j] = df[j].str.strip('')
  if (df[j].map({'yes':1,'no':0}).isna().sum() / (df.size/len(df.columns))) <= 0.05:
    for i in df.index:
      if df[j][i] == '':
        df = df.drop(i,axis='index')
  df[j] = df[j].map({'yes':1,'no':0})
```

```python
df.info()
```

#ðŸ“Š Carga y anÃ¡lisis

```python
df.describe()
```

```python
dicc = {'Situacion':[int((df.size/len(df.columns))-(df['Churn'].sum())),int(df['Churn'].sum())]}
df_churn = pd.DataFrame(data=dicc, index=['No Churn', 'Churn'])
```

```python
#General
fig, ax = plt.subplots(figsize=(8,4))
ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y',which='both',size=0)
ax = sns.barplot(data=df_churn,x='Situacion',y=df_churn.index,hue=df_churn.index,orient = 'h',
                  palette='Paired',legend=False)
ax.set_title(label='SituaciÃ³n de abandono de clientes',
      loc='left',fontsize=16)
ax.set_ylabel('SituaciÃ³n')
for i,j in enumerate(df_churn['Situacion']):
  ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')
plt.show()
```

```python
dicc_2 = {'Situacion_por_Genero':[0,0,0,0]}
#Hombres vs Mujeres
for i in df.index:
  if df['gender'][i] == 'Male' and df['Churn'][i] == 0:
    dicc_2['Situacion_por_Genero'][0] += 1
  elif df['gender'][i] == 'Male' and df['Churn'][i] == 1:
    dicc_2['Situacion_por_Genero'][1] += 1
  elif df['gender'][i] == 'Female' and df['Churn'][i] == 0:
    dicc_2['Situacion_por_Genero'][2] += 1
  elif df['gender'][i] == 'Female' and df['Churn'][i] == 1:
    dicc_2['Situacion_por_Genero'][3] += 1

indice = ['Estancia de Hombres', 'EvasiÃ³n de Hombres', 'Estancia de Mujeres', 'EvasiÃ³n de Mujeres']
df_churn_por_genero = pd.DataFrame(data=dicc_2, index=indice)
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y',which='both',size=0)
ax = sns.barplot(data=df_churn_por_genero,x='Situacion_por_Genero',y=df_churn_por_genero.index,hue=df_churn_por_genero.index,orient = 'h',
                  palette='Paired',legend=False)
ax.set_title(label='SituaciÃ³n de abandono de clientes por gÃ©nero',
      loc='left',fontsize=16)
ax.set_ylabel('SituaciÃ³n por genero')
for i,j in enumerate(df_churn_por_genero['Situacion_por_Genero']):
  ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')
plt.show()
```

```python
senior_churn = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='total')
senior_churn['SeniorCitizen']=senior_churn['SeniorCitizen'].map({0:'No Senior',1:'Senior'})
senior_churn['Churn']=senior_churn['Churn'].map({0:'Se mantiene',1:'AbandonÃ³'})

sns.barplot(data=senior_churn,x='SeniorCitizen',y='total',hue='Churn',palette='Paired')

plt.title('Cantidad de abandono segÃºn si es Senior Citizen o no', loc='left', fontsize=14)
plt.xlabel('Senior Citizen')
plt.ylabel('Cantidad de clientes')

plt.show()
```

```python
churn_contrato = df.groupby(['Contract', 'Churn']).size().reset_index(name='total')
churn_contrato['Churn']=churn_contrato['Churn'].map({0:'Se mantiene',1:'AbandonÃ³'})

sns.barplot(data=churn_contrato,x='Contract',y='total',hue='Churn',palette='Paired')

plt.title('Cantidad de abandono segÃºn tipo de contrato', loc='left', fontsize=14)
plt.xlabel('Tipo de contrato')
plt.ylabel('Cantidad de clientes')

plt.show()
```

```python
churn_pareja = df.groupby(['Partner', 'Churn']).size().reset_index(name='total')
churn_pareja['Partner']=churn_pareja['Partner'].map({0:'Sin pareja',1:'Con pareja'})
churn_pareja['Churn']=churn_pareja['Churn'].map({0:'Se mantiene',1:'AbandonÃ³'})

sns.barplot(data=churn_pareja,x='Partner',y='total',hue='Churn',palette='Paired')

plt.title('Cantidad de abandono segÃºn pareja', loc='left', fontsize=14)
plt.xlabel('Tipo de contrato')
plt.ylabel('Cantidad de clientes')

plt.show()
```

```python
churn_metodo_pago = df.groupby(['PaymentMethod', 'Churn']).size().reset_index(name='total')
churn_metodo_pago['Churn']=churn_metodo_pago['Churn'].map({0:'Se mantiene',1:'AbandonÃ³'})

sns.barplot(data=churn_metodo_pago,x='total',y='PaymentMethod',hue='Churn',palette='Paired',orient='h')

plt.title('Cantidad de abandono segÃºn metodo de pago', loc='left', fontsize=14)
plt.xlabel('Total de clientes')
plt.ylabel('MÃ©todo de pago')

plt.show()
```

```python
df = df.rename(columns={'Charges.Total':'Cargos Totales'})

churn_total_gastado = df.groupby(['Cargos Totales', 'Churn']).size().reset_index(name='total')
churn_total_gastado['Churn']=churn_total_gastado['Churn'].map({0:'Se mantiene',1:'AbandonÃ³'})

sns.boxplot(x='Churn', y='Cargos Totales', data=churn_total_gastado)
plt.title('DistribuciÃ³n del Abandono segÃºn Total Gastado')
plt.show()
```

```python
df = df.rename(columns={'Charges.Monthly':'Cargos Mensuales'})

churn_total_gastado = df.groupby(['Cargos Mensuales', 'Churn']).size().reset_index(name='total')
churn_total_gastado['Churn']=churn_total_gastado['Churn'].map({0:'Se mantiene',1:'AbandonÃ³'})

sns.boxplot(x='Churn', y='Cargos Mensuales', data=churn_total_gastado)
plt.title('DistribuciÃ³n del Abandono segÃºn Gasto Mensual')
plt.show()
```

```python
df = df.rename(columns={'Cuentas_Diarias':'Cargos Diarios'})

churn_total_gastado = df.groupby(['Cargos Diarios', 'Churn']).size().reset_index(name='total')
churn_total_gastado['Churn']=churn_total_gastado['Churn'].map({0:'Se mantiene',1:'AbandonÃ³'})

sns.boxplot(x='Churn', y='Cargos Diarios', data=churn_total_gastado)
plt.title('DistribuciÃ³n del Abandono segÃºn Gasto Diario')
plt.show()
```
